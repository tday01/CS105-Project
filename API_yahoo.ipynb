```python
# list all ticker symbols plus index symbol (ndx): 103 + 1
names_ndx_df = pd.read_csv('./data_files/stock_name.txt')

# get stock names from txt file
f = open('./data_files/stock_name.txt', 'r')
stocks_list = f.read().splitlines()
f.close()

# choose stocks
# yf_nasdaq_104 = YahooFinancials(stocks_list)
yf_nasdaq_104 = YahooFinancials(['AAL','AAPL','ADBE'])

# pull data from API
stock_data = yf_nasdaq_104.get_stock_price_data()
#stock_data = yf_nasdaq_104.get_summary_data()
#stock_data = yf_nasdaq_104.get_historical_stock_data('2020-1-12','2020-2-12','daily')

# save json stock data
with open('./data_files/yf_nasdaq_104.json', 'w') as outfile:
    json.dump(stock_data, outfile)

# read json file into dataframe and transpose rows/cols
df = pd.read_json('./data_files/yf_nasdaq_104.json').T

# save a csv version of stock data
df.to_csv('./data_files/yf_nasdaq_104.csv')

df.head(3)
```
